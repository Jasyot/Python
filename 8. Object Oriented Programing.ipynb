```python
#defining a class

class MyClass:
    
    def __init__(self): #will discuss later!! --------- LINE1
        self.variable1_my_class = 200 #class variables
        self.variable2_my_class = 300 #class variables

    def my_class_func(self): #class function/method/attribute
        print("Hello from my_class_func")

#creation of class object called MyClass()
# Creating an instance object of MyClass called my_obj
my_obj = MyClass()

# Calling a method of the class on the instance object
my_obj.my_class_func() 

# Accessing variables of the class using the instance object
print(my_obj.variable1_my_class)
print(my_obj.variable2_my_class)
```

```python
#Example:

class MyClass:
    # A class object variable
    class_attribute = "I am a class attribute"

    def __init__(self, value):
        # An instance object variable
        self.instance_attribute = value

    # Instance method
    # Uses 'self' to refer to the individual instance of the class.
    # Can access and modify object attributes and other instance methods.
    def instance_method(self):
        print(f"This is an instance method. Instance attribute: {self.instance_attribute}")

    # Class method
    # Uses 'cls' to refer to the class itself, not the instance.
    # Can modify class state that applies across all instances of the class.
    @classmethod
    def class_method(cls):
        print(f"This is a class method. Class attribute: {cls.class_attribute}")

    # Static method
    # Does not take 'self' or 'cls' parameters.
    # Behaves like a regular function but belongs to the class's namespace.
    @staticmethod
    def static_method():
        print("This is a static method.")

# Creating an instance of MyClass
my_obj = MyClass("Instance value")

# Calling instance method
my_obj.instance_method()

# Calling class method
MyClass.class_method()

# Calling static method
MyClass.static_method()
```
