```python
#Example1:

def func(n):
    
    if n==1:
        return 1
    
    else:
        s=n+func(n-1) #Recursion occuring as func(n-1) is called with in the func(n)!!
        return s
```

```python
func(3)
```

```python
#Without recursion

def sum_natural():
    
    n=int(input("till what numb sum needed."))
    mylist=list(range(1,n+1))
    print('the list of number is',mylist)
    
    sum_list=sum(mylist)
    return sum_list
```

```python
sum_natural()
```

```python
#Using recursion:

n=int(input('enter the till where the sum is needed'))

def sum_rec(n):
    
    if n==1:
        return 1
    else:
        s=n + sum_rec(n-1)
        return s

sum_rec(n)
```
